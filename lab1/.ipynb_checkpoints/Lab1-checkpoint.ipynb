```python
import os
import regex
import matplotlib.pyplot as plt
```

```python
files_names = []
for file in os.listdir('./ustawy'):
    files_names.append(os.path.join("ustawy",file))
```

```python
unit_results = {}
ustawa_occurs = 0
ustawa_followed_z_dnia_occurs = 0
ustawa_not_followed_z_dnia_occurs = 0
ustawa_excluding_o_zmianie = 0

unit_add = regex.compile('dodaje się (?:ust|pkt|art|§)', regex.IGNORECASE)
unit_del = regex.compile('skreśla się (?:ust|pkt|art|§)', regex.IGNORECASE)            
unit_mod = regex.compile('(?:pkt*|art*|ust*|§*) [0-9]* otrzymuje brzmienie', regex.IGNORECASE)

ustawa_regex = regex.compile('(?:\\bustawa\\b|\\bustawy\\b|\\bustawie\\b|\\bustawę\\b' \
                                 '|\\bustawą\\b|\\bustawo\\b|\\bustawy\\b|\\bustaw\\b' \
                                 '|\\bustawom\\b|\\bustawami\\b|\\bustawach\\b)',
                                regex.IGNORECASE)
ustawa_followed_regex = regex.compile('(?:\\bustawa\\b|\\bustawy\\b|\\bustawie\\b|\\bustawę\\b' \
                                 '|\\bustawą\\b|\\bustawo\\b|\\bustawy\\b|\\bustaw\\b' \
                                 '|\\bustawom\\b|\\bustawami\\b|\\bustawach\\b)(?= z dnia)',
                                regex.IGNORECASE)
ustawa_not_followed_regex = regex.compile('(?:\\bustawa\\b|\\bustawy\\b|\\bustawie\\b|\\bustawę\\b' \
                                 '|\\bustawą\\b|\\bustawo\\b|\\bustawy\\b|\\bustaw\\b' \
                                 '|\\bustawom\\b|\\bustawami\\b|\\bustawach\\b)(?! z dnia)',
                                regex.IGNORECASE)
ustawa_excluding_regex = regex.compile('(?:\\bustawa\\b|\\bustawy\\b|\\bustawie\\b|\\bustawę\\b' \
                                 '|\\bustawą\\b|\\bustawo\\b|\\bustawy\\b|\\bustaw\\b' \
                                 '|\\bustawom\\b|\\bustawami\\b|\\bustawach\\b)(?! o zmianie)',
                                regex.IGNORECASE)

for file_name in files_names:
    
    year = int(file_name[7:11])
    
    if year not in unit_results.keys():
        unit_results[year] = {}
        unit_results[year]['add'] = 0
        unit_results[year]['mod'] = 0
        unit_results[year]['del'] = 0
    
    with open(file_name) as file:

        for line in file:
            
            add_matches = unit_add.findall(line)
            del_matches = unit_del.findall(line)
            mod_matches = unit_mod.findall(line)
            
            ustawa_matches = ustawa_regex.findall(line)
            ustawa_followed_matches = ustawa_followed_regex.findall(line)
            ustawa_not_followed_matches = ustawa_not_followed_regex.findall(line)
            ustawa_excluding_matches = ustawa_excluding_regex.findall(line)

            unit_results[year]['add'] += len(add_matches)
            unit_results[year]['del'] += len(del_matches)
            unit_results[year]['mod'] += len(mod_matches)
            
            ustawa_occurs += len(ustawa_matches)
            ustawa_followed_z_dnia_occurs += len(ustawa_followed_matches)
            ustawa_not_followed_z_dnia_occurs += len(ustawa_not_followed_matches)
            ustawa_excluding_o_zmianie += len(ustawa_excluding_matches)

unit_results = dict(sorted(unit_results.items()))
print(unit_results)
```

Wynikiem jest liczba każdego typu poprawek w każdym roku. 

```python
fig_year_changes = plt.figure()
ax = plt.axes()
X = list(unit_results.keys())
addY = [val['add'] for val in unit_results.values()]
delY = [val['del'] for val in unit_results.values()]
modY = [val['mod'] for val in unit_results.values()]

ax.plot(X, addY, color='green', label='added')
ax.plot(X, delY, color='red', label='deleted')
ax.plot(X, modY, color='blue', label='changed')

ax.legend()
plt.title('Amount of added, removed and changed units per year')

plt.show()
```

Na wykresie liczby danych poprawek w każdym roku można zauważyć, że zmian było najwięcej, a usunięć najmniej.

```python
summed_add = sum([val['add'] for val in unit_results.values()])
summed_del = sum([val['del'] for val in unit_results.values()])
summed_mod = sum([val['mod'] for val in unit_results.values()])

perc_add = [val['add']/summed_add*100 for val in unit_results.values()]
perc_del = [val['del']/summed_del*100 for val in unit_results.values()]
perc_mod = [val['mod']/summed_mod*100 for val in unit_results.values()]

fig_year_perc = plt.figure()
ax = plt.axes()
X = list(unit_results.keys())

ax.plot(X, perc_add, color='green', label='added [%]')
ax.plot(X, perc_del, color='red', label='deleted [%]')
ax.plot(X, perc_mod, color='blue', label='changed [%]')

ax.legend()
plt.title('Percentage of added, removed and changed units per year')

plt.show()
```

Można zauważyć, że najwięcej usunięć było wprowadzonych w roku 1997 oraz 2000. Natomiast w roku 2002 wszystkich poprawek było mniej. Aż w 2003 podjęto znowu zmiany i dodawanie.

```python
print(ustawa_occurs)
print(ustawa_followed_z_dnia_occurs)
print(ustawa_not_followed_z_dnia_occurs)

ustawa_occurs == ustawa_followed_z_dnia_occurs+ustawa_not_followed_z_dnia_occurs
```

Liczba wystąpień wyrazu ustawa bez następującego 'z dnia' oraz z występującym sumuje się do liczby wszystkich wystąpień tego wyrazu.

```python
print(ustawa_excluding_o_zmianie)
```
